##### **Q1. 데이터를 분리하고 난 후 검증 세트(X_val)의 개수를 제출하세요.**

```python
# 데이터셋을 불러옵니다. 

import pandas as pd
import numpy as np

df = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kc_house_data/kc_house_data.csv')
df
```

```python
from sklearn.model_selection import train_test_split

feature = df[['bedrooms', 'bathrooms', 'sqft_living']]
target = df['price']


X_train, X_test, y_train, y_test  = train_test_split(feature, target, test_size = 0.2, random_state = 2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 2)


print("validation set : ", X_val.shape)
print("X_train",X_train.shape)
print("X_test",X_test.shape)
```

##### **Q2. 기준모델과 다중선형회귀 모델의 성능으로 옳은 것을 고르세요. (반올림하여 소수점 둘째 자리까지)**
- 모델을 만들기 전에 훈련 데이터 타겟의 평균값을 사용한 기준모델을 만들고, 기준모델의 RMSE, $R^2$ score를 확인해보세요.
- 다중선형회귀 모델을 만들고 훈련셋과 검증셋의 RMSE, $R^2$ score를 비교해보세요.
- 마지막으로 테스트셋을 사용해서 모델의 일반화 성능을 확인해보세요.
  - **퀴즈 보기**에 나온 다중선형회귀 모델의 score는 **테스트셋을 사용한 값**입니다.
  - 성능은 RMSE, $R^2$ score를 사용하세요.
- MSE, MAE도 확인해보세요.

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from IPython.display import display, HTML
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:,.2f}'.format



def eval_models(y_true, y_pred) :
    mse = mean_squared_error(y_true, y_pred).round(3)
    rmse = np.sqrt(mse).round(3)
    mae = mean_absolute_error(y_true, y_pred).round(3)
    r2 = r2_score(y_true, y_pred)

    return mse, rmse, mae, r2

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)
y_test_pred = model.predict(X_test)

baseline = [y_train.mean()] * len(y_train)

comparison = pd.DataFrame(index=['mse','rmse', 'mae', 'r2'], columns=['Base', 'Train', 'Validation', 'Test'])
comparison['Base'] = eval_models(y_train, baseline)
comparison['Train'] = eval_models(y_train, y_train_pred)
comparison['Validation'] = eval_models(y_val, y_val_pred)
comparison['Test'] = eval_models(y_test, y_test_pred)

display(comparison)


```

##### **Q3. cross_val_score 함수에 대한 설명으로 옳은 것을 고르세요.**

```python
from sklearn.model_selection import KFold

target = 'price'
features = ['bedrooms', 'bathrooms', 'sqft_living']

def x_y_split(df):
  X = df[features]
  y = df[target]
  return X, y

df_train, df_test = train_test_split(df, test_size = 0.2, random_state = 2)
X_train, y_train = x_y_split(df_train)
X_test, y_test = x_y_split(df_test)

kf = KFold(n_splits=10)
print('k =',kf.get_n_splits())

cv_result = []
for train_idx, test_idx in kf.split(X_train) :
    X_train_cv, X_val_cv = X_train.iloc[train_idx], X_train.iloc[test_idx]
    y_train_cv, y_val_cv = y_train.iloc[train_idx], y_train.iloc[test_idx]

    model.fit(X_train_cv, y_train_cv)
    y_pred_cv = model.predict(X_val_cv)
    
    mse = mean_squared_error(y_val_cv, y_pred_cv).round(3)
    rmse = np.sqrt(mse).round(3)
    cv_result.append(rmse)

print('rmse for each fold : ', cv_result)
print('average mae for model : ', np.mean(cv_result).round(2))
print('std of result : ', np.std(cv_result).round(2))
```

##### **Q4. 다항선형회귀를 만들고 최적의 차수를 찾아 제출하세요.**

```python
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

feature = df[['bedrooms', 'bathrooms', 'sqft_living']]
target = df['price']

def PolynomialRegression(degree=3, **kwargs):
    return make_pipeline(PolynomialFeatures(degree), 
                         LinearRegression(**kwargs))


X_train, X_test, y_train, y_test  = train_test_split(feature, target, test_size = 0.2, random_state = 2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 2)


degrees = [1, 2, 3, 4, 5]

best_degree = None

for i in degrees:
  poly_model = PolynomialRegression(i)
  poly_model.fit(X_train, y_train)
  y_pred_val = poly_model.predict(X_val)
  print("차수",i)
  print(eval_models(y_val, y_pred_val))

```

```python
for i in degrees:
  poly_model = PolynomialRegression(i)
  poly_model.fit(X_train, y_train)
  y_predict_train = poly_model.predict(X_train)
  print("train 데이터 차수",i)
  print(eval_models(y_train, y_predict_train))

print('\n')
for i in degrees:
  poly_model = PolynomialRegression(i)
  poly_model.fit(X_train, y_train)
  y_predict_test = poly_model.predict(X_test)
  print("test 데이터 차수",i)
  print(eval_models(y_test, y_predict_test))  
```
